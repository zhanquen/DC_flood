```python
import meshio

# File path to your XDMF file
xdmf_path = "/Users/aureliengirardot/Documents/Aurelien/mines/Data Challenge fluides/4Students_AnXplore03/AllFields_Resultats_MESH_1.xdmf"

# Open the XDMF file
reader = meshio.xdmf.TimeSeriesReader(xdmf_path)

# Read mesh points and cells (mesh geometry)
points, cells = reader.read_points_cells()

# Extract data for each timestep
meshes = []
print("reader.num_steps", reader.num_steps)
for i in range(reader.num_steps):
    try:
        time, point_data, cell_data, _ = reader.read_data(i)
    except ValueError:
        time, point_data, cell_data = reader.read_data(i)
    
    # Create a Mesh object
    mesh = meshio.Mesh(points, cells, point_data=point_data, cell_data=cell_data)
    meshes.append(mesh)

# Access the mesh data
print(f"Loaded {len(meshes)} timesteps")
print(meshes)
```

```python
# Loop through each timestep's mesh and print the Vitesse field
for i, mesh in enumerate(meshes):
    if "Vitesse" in mesh.point_data:
        print(f"Timestep {i}: Vitesse field shape: {mesh.point_data['Vitesse'].shape}")
        print(f"Timestep {i}: First 5 Vitesse values:\n{mesh.point_data['Vitesse'][:5]}")
    else:
        print(f"Timestep {i}: Vitesse field not found.")
```

```python
# Loop through each timestep's mesh and print the Pression field
for i, mesh in enumerate(meshes):
    if "Pression" in mesh.point_data:
        print(f"Timestep {i}: Pression field shape: {mesh.point_data['Pression'].shape}")
        print(f"Timestep {i}: First 5 Pression values:\n{mesh.point_data['Pression'][:5]}")
    else:
        print(f"Timestep {i}: Pression field not found.")
```
